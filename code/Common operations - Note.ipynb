```python
import pandas as pd

heart = pd.read_csv('../data/Heart.csv')
heart.head(3)
```

```python
# checck number of NaN
# print(heart.isnull().values.any())
heart.isnull().sum()
```

```python
heart.describe()
```

```python
# fillna with mean
mean_missing = heart.Ca.mean()
heart.Ca = heart.Ca.fillna(mean_missing)

heart.isnull().sum()
```

```python
# drop unwanted columns
heart = heart.drop(heart.columns[0], axis=1)
```

```python
# heart.ChestPain = heart.ChestPain.map({'typical': 0, 'asymptomatic': 1})
heart = pd.get_dummies(heart, columns=['ChestPain', 'Thal'], prefix=['ChestPain', 'Thal'])
heart = heart.drop([heart.columns[0], 'ChestPain_typical', 'Thal_normal'], axis=1)
```
